## Space and span

The space spanned by vectors $\big\{ \mathbf{v}_{1}, \mathbf{v}_{2}, \ldots, \mathbf{v}_n \big\} \equiv$ the space of vectors in the form of: $\mathbf{v} = \alpha_{1} \mathbf{v}_{1} + \alpha_{2} \mathbf{v}_{2} + \ldots + \alpha_{n} \mathbf{v}_{n}$ where $\alpha \in \mathbb{R}$.

### Example

The space spanned by $\mathbf{v}_{1} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\mathbf{v}_{2} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ is $\mathbb{R}^{2}$.

### Example

The space spanned by $\mathbf{v}_{1} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ and $\mathbf{v}_{2} = \begin{bmatrix} -1 \\ 3 \end{bmatrix}$ is $\mathbb{R}^{2}$.

### Example

The space spanned by $\mathbf{v}_{1} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$ and $\mathbf{v}_{2} = \begin{bmatrix} -4 \\ -8 \end{bmatrix}$ is a line.


## Projectors

Given $\mathbf{u}, \mathbf{v} \in \mathbb{R}^{n}$, a projector is defined as $\mathbf{P}$ where:

$$
\mathbf{u} = \mathbf{P}\mathbf{v}
$$

![orthogonal_projector](../lecture_diagrams/projector_1.jpg)

Projectors are $n \times n$ matrices and the following relation holds true:

$$
\mathbf{P}(\mathbf{Pv}) = \mathbf{Pv}
$$

Which means that:

$$
\begin{align}
  \mathbf{P}(\mathbf{Pv}) & = (\mathbf{PP})\mathbf{v} \\
  & = \mathbf{P}^{2} \mathbf{v} \\
  & = \mathbf{Pv} \quad \implies \mathbf{P}^{2} = \mathbf{P}
\end{align}
$$

The matrix $\mathbf{P}\rvert_{n\times n}$ is a projector if $\mathbf{P}^{2} = \mathbf{P}$.

### Example

If $\mathbf{U}\rvert_{n\times r} = \big[ \mathbf{u}_{1}, \mathbf{u}_{2}, \ldots, \mathbf{u}_{r} \big]$ where $n \gt r$ and columns of $\mathbf{U}$ are a set of orthonormal vectors, then $\mathbf{P} = \mathbf{U} \mathbf{U}^{T}$ is a projector.

For orthonormal vectors:

$$
\langle \mathbf{u}_{i}, \mathbf{u}_{j} \rangle = \mathbf{u}_{i}^{T} \mathbf{u}_{j} = \delta_{ij} 
\begin{cases}
  1 & i = j \\
  0 & i \neq j
\end{cases}
$$
<br>
$$
\mathbf{U}^{T} \mathbf{U} = 
\begin{bmatrix}
  \mathbf{u}_{1}^{T} \\
  \mathbf{u}_{2}^{T} \\
  \vdots \\
  \mathbf{u}_{r}^{T}
\end{bmatrix}
\begin{bmatrix}
  \mathbf{u}_{1} & \mathbf{u}_{2} & \ldots & \mathbf{u}_{r} 
\end{bmatrix} = 
\begin{bmatrix}
  \mathbf{u}_{1}^{T} \mathbf{u}_{1} & \mathbf{u}_{1}^{T} \mathbf{u}_{2} & \ldots & \mathbf{u}_{1}^{T} \mathbf{u}_{r} \\
  \mathbf{u}_{2}^{T} \mathbf{u}_{1} & \mathbf{u}_{2}^{T} \mathbf{u}_{2} & \ldots & \mathbf{u}_{2}^{T} \mathbf{u}_{r} \\
  \vdots & \vdots & \ddots & \vdots \\
  \mathbf{u}_{r}^{T} \mathbf{u}_{1} & \mathbf{u}_{r}^{T} \mathbf{u}_{2} & \ldots & \mathbf{u}_{r}^{T} \mathbf{u}_{r}
\end{bmatrix} =
\mathbf{I}\rvert_{r \times r}
$$

Given this suppose $n = 1000$ and $r = 5$ where $\mathbf{P} \equiv (\mathbf{UU}^{T})_{n \times n}$.  Then:

$$
\begin{align}
  \mathbf{P}^{2} & = (\mathbf{UU}^{T})(\mathbf{UU}^{T}) \\
  & = \mathbf{UU}^{T} \\
  & = \mathbf{P}
\end{align}
$$


## Complementary projector

The complemetary projection is defined as $\mathbf{P}' = \mathbf{I} - \mathbf{P}$.

$$
\begin{align}
  (\mathbf{I} - \mathbf{P})^{2} & = \\
  & = \mathbf{I} -2\mathbf{P} + \mathbf{P}^{2} \\
  & = \mathbf{I} -2\mathbf{P} + \mathbf{P} \\
  & = \mathbf{I} - \mathbf{P}
\end{align}
$$


## Orthogonal projectors

$\mathbf{P}_{n \times n}$ is an orthogonal projector if $\begin{cases} \mathbf{P}^{2} = \mathbf{P} \\ \mathbf{P}^{T} = \mathbf{P} \end{cases}$

### Example

$$
\langle \mathbf{Pv}, (\mathbf{I} - \mathbf{P}) \mathbf{v}) = 0
$$

#### Proof

$$
\begin{align}
  \langle \mathbf{Pv}, (\mathbf{I} - \mathbf{P}) \mathbf{v} \rangle & = 0 \\ \\
  (\mathbf{Pv})^{T}((\mathbf{I} - \mathbf{P}) \mathbf{v}) & = 0  \\ \\
  (\mathbf{Pv})^{T}(\mathbf{I} - \mathbf{P})\mathbf{v} & = 0 \\ \\
  \mathbf{v}^{T} \mathbf{P}^{T}(\mathbf{I} - \mathbf{P})\mathbf{v} & = 0 \\ \\
  \mathbf{v}^{T} \mathbf{P}^{T} \mathbf{I} \mathbf{v} - \mathbf{v}^{T} \mathbf{P}^{T} \mathbf{P} \mathbf{v} & = 0 \\ \\
  \mathbf{v}^{T} \mathbf{P}^{T} \mathbf{v} - \mathbf{v}^{T} \mathbf{P}^{T} \mathbf{P} \mathbf{v} & = 0 \\ \\
  \mathbf{v}^{T}(\mathbf{P}^{T} - \mathbf{P}^{T} \mathbf{P})\mathbf{v} & = 0 \\ \\
  \implies \mathbf{P}^{T} - \mathbf{P}^{T} \mathbf{P} & = 0 
  \iff \mathbf{P}^{T} = \mathbf{P} \wedge \mathbf{P}^{2} = \mathbf{P}
\end{align}
$$


## Singular value decomposition

$$
\mathbf{A}_{n \times m} = \mathbf{U}_{n \times m} \mathbf{\Sigma}_{m \times m} \mathbf{V}^{T}_{m \times m}
$$

- Left singular vectors: $\mathbf{U}^{T} \mathbf{U} = \mathbf{I}_{m \times m}$ and $\mathbf{U}\text{: orthonormal}$.

- Right singular values: $\mathbf{V} \mathbf{V} = \mathbf{I}_{m \times m}$ and $\mathbf{V}\text{: orthonormal}$.

- Singular values: $\mathbf{\Sigma} = \begin{bmatrix} \sigma_{1} & 0 & 0 & \ldots & 0 \\ 0 & \sigma_{2} & 0 & \ldots & \vdots \\ 0 & 0 & \ddots & \ldots & \vdots \\ 0 & 0 & 0 & \ddots & \vdots \\ 0 & \ldots & \ldots & \sigma_{m} & 0  \end{bmatrix}$, and $\sigma_{1} \geqslant \sigma_{2} \geqslant \ldots \geqslant \sigma_{m}$
