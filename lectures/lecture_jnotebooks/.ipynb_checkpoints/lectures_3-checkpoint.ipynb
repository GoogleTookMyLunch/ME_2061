## Least squares

$$
\Bigg[ \mathbf{f} \Bigg]_{n \times 1} \simeq \Bigg[ \mathbf{u}_{1} \Bigg]^{\alpha_{1}}_{n \times 1} + \Bigg[ \mathbf{u}_{2} \Bigg]^{\alpha_{2}}_{n \times 1} + \ldots + \Bigg[ \mathbf{u}_{r} \Bigg]^{\alpha_{r}}_{n \times 1}
$$

where: 

- $n$: ambient dimension

- $r$: the number of basis vectors

- $r \lt n$

- $\big\{ \mathbf{u}_{i} \big\}_{i = 1}^{r}$: independent vectors

$$
\mathbf{f} \simeq \mathbf{\hat{f}} = \sum_{i = 1}^{r}\alpha_{i} \mathbf{u}_{i} = \mathbf{U} \mathbf{\alpha}
\qquad
\mathbf{U} = \big[ \mathbf{u}_{1}, \mathbf{u}_{2}, \ldots, \mathbf{u}_{r} \big]_{n \times r}
\qquad
\mathbf{\alpha} = 
\begin{bmatrix}
  \alpha_{1} \\
  \alpha_{2} \\
  \vdots \\
  \alpha_{r}
\end{bmatrix}
$$
<br>
$$
\mathbf{f}_{n \times 1} = \mathbf{\hat{f}}_{n \times 1} + \mathbf{e}_{n \times 1} \implies \mathbf{e} = \mathbf{f} - \mathbf{U \alpha}
$$
<br>
$$
\begin{align}
  \mathrm{E}(\alpha) = \big\lVert \mathbf{e} \big\rVert^{2} & = \big\lVert
  \mathbf{f} - \mathbf{U \alpha} \big\rVert^{2} \\ \\
  & = \big( \mathbf{f} - \mathbf{U \alpha} \big)^{T} \big( \mathbf{f} - \mathbf{U \alpha} \big) \\ \\
  & = \big( \mathbf{f}^{T} - \big( \mathbf{U \alpha} \big)^{T} \big) \big( \mathbf{f} - \mathbf{U \alpha} \big) \\ \\
  & = \big( \mathbf{f}^{T} - \mathbf{\alpha}^{T} \mathbf{U}^{T} \big) \big( \mathbf{f} - \mathbf{U \alpha} \big) \\ \\
  & = \mathbf{f}^{T} \mathbf{f} - \mathbf{f}^{T} \mathbf{U \alpha} - \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{f} + \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{U} \mathbf{\alpha} \\ \\
  & = \mathbf{f}^{T} \mathbf{f} - \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{f} - \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{f} + \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{U} \mathbf{\alpha} \\ \\
  & = \mathbf{f}^{T} \mathbf{f} - 2\mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{f} + \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{U} \mathbf{\alpha}
\end{align}
$$

The best $\alpha$: 

$$
\frac{\partial \mathrm{E}}{\partial \mathbf{\alpha}} = 
\begin{bmatrix}
  \frac{\partial \mathrm{E}}{\partial \alpha_{1}} \\
  \frac{\partial \mathrm{E}}{\partial \alpha_{2}} \\
  \vdots \\
  \frac{\partial \mathrm{E}}{\partial \alpha_{r}}
\end{bmatrix}
$$

Note that:

$$
\frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{c} \big)}{\partial \alpha} = 
\frac{\partial \big( \alpha_{1} c_1 + \alpha_{2} c_2 + \ldots \alpha_{r} c_{r} \big)}{\partial \alpha} =
\begin{bmatrix}
  \frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{c} \big)}{\partial \alpha_{1}} \\
  \frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{c} \big)}{\partial \alpha_{}} \\
  \vdots \\
  \frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{c} \big)}{\partial \alpha_{r}}
\end{bmatrix} = 
\begin{bmatrix}
  c_1 \\
  c_2 \\
  \vdots \\
  c_r
\end{bmatrix} = 
\mathbf{c}
$$

Therefore:

$$
\frac{\partial \big( \mathbf{f}^{T} \mathbf{f} \big)}{\partial \mathbf{\alpha}} =
\mathbf{0}_{r \times 1}
\qquad
\frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{f} \big)}{\partial \mathbf{\alpha}} = 
\big( \mathbf{U}^{T} \mathbf{f} \big)_{r \times 1}
\qquad
\frac{\partial \big( \mathbf{\alpha}^{T} \mathbf{U}^{T} \mathbf{U} \alpha \big)}{\partial \alpha} =
2 \mathbf{U}^{T} \mathbf{U} \mathbf{\alpha}
$$

and:

$$
\frac{\partial \mathrm{E}}{\partial \mathbf{\alpha}} = -2 \mathbf{U}^{T} \mathbf{f} + 2\mathbf{U}^{T} \mathbf{U} \mathbf{\alpha} = \mathbf{0}
\qquad
\implies
\qquad
\mathbf{U}^{T} \mathbf{U} \mathbf{\alpha} = \mathbf{U}^{T} \mathbf{f}
\qquad
\implies
\qquad
\mathbf{\alpha} = \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \mathbf{f}
$$
<br>
$$
\mathbf{\hat{f}} = \mathbf{U} \mathbf{\alpha} = \mathbf{U} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \mathbf{f}
$$
<br>
$$
\mathcal{P}_{n \times n} = \big[ \mathbf{U} \big]_{n \times r} \big[ \mathbf{U}^{T} \mathbf{U} \big]^{-1}_{r \times r} \big[ \mathbf{U}^{T} \big]_{r \times n}
$$

Check that $\mathcal{P}$ is a projector:

$$
\begin{align}
  \mathcal{P}^{2} & = \big( \mathbf{U} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \big) \big( \mathbf{U} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1}
  \mathbf{U}^{T} \big) \\ \\
  & = \mathbf{U} \mathbf{I} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \\ \\
  & = \mathbf{U} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \\ \\ 
  & = \mathcal{P}
\end{align}
$$

Check if $\mathcal{P}$ is an orthogonal projector (homework problem):

$$
\mathcal{P}^{T} = \mathcal{P}
$$

$\mathcal{P}$ is an orthogonal projector to the space spanned by $\mathbf{U}$.

## Singular value decomposition

$$
\mathbf{A}_{n \times m} = \mathbf{U}_{n \times m} \mathbf{\Sigma}_{m \times m} \mathbf{V}^{T}_{m \times m}
$$

- Left singular vectors: $\mathbf{U}^{T} \mathbf{U} = \mathbf{I}_{m \times m}$ and $\mathbf{U}\text{: orthonormal}$.

- Right singular values: $\mathbf{V} \mathbf{V} = \mathbf{I}_{m \times m}$ and $\mathbf{V}\text{: orthonormal}$.

- Singular values: $\mathbf{\Sigma} = \begin{bmatrix} \sigma_{1} & 0 & 0 & \ldots & 0 \\ 0 & \sigma_{2} & 0 & \ldots & \vdots \\ 0 & 0 & \ddots & \ldots & \vdots \\ 0 & 0 & 0 & \ddots & \vdots \\ 0 & \ldots & \ldots & \sigma_{m} & 0  \end{bmatrix}$, and $\sigma_{1} \geqslant \sigma_{2} \geqslant \ldots \geqslant \sigma_{m}$
