### Problem 1

Show that the projector $\mathbf{P} \in \mathbb{R}^{n\times n}$ is an orthogonal projector if $\mathbf{P}^{T} = \mathbf{P}$.


#### Problem 1 calculations

I need to show that the Euclidean inner product of the projector ($\mathbf{P}$) and complementary projector ($\mathbf{I} - \mathbf{P}$) are orthogonal (i.e., equal to 0).  Let $\mathbf{v} \in \mathbf{R}^{n\times 1}$ and $\mathbf{I} \in \mathbb{R}^{n\times n}$.

Show $(\mathbf{Pv})^{T}((\mathbf{I} - \mathbf{P})\mathbf{v}) = \mathbf{0}$ is true if and only if $\mathbf P$ is symmetric, that is, $\mathbf{P}^{T} = \mathbf{P}$.  If this is true, then the projector and the complementary projector are orthogonal.

Expanding the expression and using the tranpose of a product property gives:

$$
\begin{align}
  0 & = (\mathbf{Pv})^{T}((\mathbf{I} - \mathbf{P})\mathbf{v}) \\
  & = (\mathbf{Pv})^{T}(\mathbf{I}\mathbf{v} - \mathbf{P}\mathbf{v}) \\
  & = (\mathbf{Pv})^{T}\mathbf{I}\mathbf{v} - (\mathbf{Pv})^{T}\mathbf{P}\mathbf{v} \\
  & = \mathbf{v}^{T}\mathbf{P}^{T}\mathbf{v} - \mathbf{v}^{T}\mathbf{P}^{T}\mathbf{P}\mathbf{v}
\end{align}
$$

If $\mathbf{P}$ is not symmetric then:

$$
\mathbf{v}^{T}\mathbf{P}^{T}\mathbf{v} \neq \mathbf{v}^{T}\mathbf{P}^{T}\mathbf{P}\mathbf{v}
$$

which violates the original equation; however, if $\mathbf P$ is symmetric then:

$$
\begin{align}
  &\mathbf{v}^{T}\mathbf{P}\mathbf{v} = \mathbf{v}^{T}\mathbf{P}\mathbf{P}\mathbf{v} \\
  &\mathbf{v}^{T}\mathbf{P}\mathbf{v} = \mathbf{v}^{T}\mathbf{P}^{2}\mathbf{v}
\end{align}
$$

provided that $\mathbf P$ is a projector then $\mathbf{P}^{2} = \mathbf P$ which leads to:

$$
\begin{align}
  &\mathbf{v}^{T}\mathbf{P}\mathbf{v} = \mathbf{v}^{T}\mathbf{P}^{2}\mathbf{v} \\
  &\mathbf{v}^{T}\mathbf{P}\mathbf{v} = \mathbf{v}^{T}\mathbf{P}\mathbf{v}
\end{align}
$$

which is true.

In summary:
<br>
<br>
<div style="border:solid; margin: auto; text-align: center; padding: 1em">
$$
\begin{align}
  0 & = (\mathbf{Pv})^{T}((\mathbf{I} - \mathbf{P})\mathbf{v}) \\
  & = (\mathbf{Pv})^{T}(\mathbf{I}\mathbf{v} - \mathbf{P}\mathbf{v}) \\
  & = (\mathbf{Pv})^{T}\mathbf{I}\mathbf{v} - (\mathbf{Pv})^{T}\mathbf{P}\mathbf{v} \\
  & = \mathbf{v}^{T}\mathbf{P}^{T}\mathbf{v} - \mathbf{v}^{T}\mathbf{P}^{T}\mathbf{P}\mathbf{v} \\
  & = \mathbf{v}^{T}\mathbf{P}\mathbf{v} - \mathbf{v}^{T}\mathbf{P}\mathbf{P}\mathbf{v} \\
  & = \mathbf{v}^{T}\mathbf{P}\mathbf{v} - \mathbf{v}^{T}\mathbf{P}^{2}\mathbf{v} \\
  & = \mathbf{v}^{T}\mathbf{P}\mathbf{v} - \mathbf{v}^{T}\mathbf{P}\mathbf{v} \\
  & = 0
\end{align}
$$
</div>

Calculate $\big( \mathbf{U}^{T} \mathbf{U} \big)^{-1}$:

The inverse of the identity matrix $\mathbf{I}$ is just $\mathbf{I}$.

Finish $\mathbf{P}$ computation:

$$
\begin{align}

\end{align}
$$

Since the columns of $\mathbf{U}$ are orthonormal and the matrix $\mathbf{U}$ is symmetric, the outer product computation simplifies. Indices that are equal (e.g., $i = j$) result in 1 and indices that are not equal (e.g., $i /neq j$) result in 0.  The indices are equal across the diagonal so:

$$
\begin{align}
  \mathbf{P} & = \mathbf{U} \big( \mathbf{U}^{T} \mathbf{U} \big)^{-1} \mathbf{U}^{T} \\
  & = \mathbf{U} \mathbf{I} \mathbf{U}^{T} \\
  & = \mathbf{U} \mathbf{U}^{T} \\
  & = \mathbf{U} \otimes \mathbf{U} \\

#### Problem 3 calculations


Two functions are orthogonal if their inner product is zero over the interval.

$$
\int^{b}_{a} u_{i}(x) u_{j}(x) dx = 0
$$

A function is normalized if:

$$
\int^{b}_{a} u_{i}(x) u_{i}(x) dx = 1
$$

```python
import numpy as np
from scipy.integrate import quad

# Define the function that does inner products
def inner_product(u1, u2):
    def integrand(x):
        return u1(x) * u2(x)
    
    # Integrate over the range [-1, 1]
    result, _ = quad(integrand, -1, 1)
    
    return result

# Define the functions
def u0(x):
    return np.sqrt(1/2)

def u1(x):
    return np.sqrt(3/2)*x

def u2(x):
    return np.sqrt(5/2)*(1/2)*(3*x**2 - 1)

def u3(x):
    return np.sqrt(7/2)*(1/2)*(5*x**3 - 3*x)
```

##### $u_{0}(x)$ and $u_{1}(x)$

```python
# Calculate the inner product of the two functions
result = inner_product(u0, u1)
print("Inner product:", result)
```
