Those were named the Four Fundamental Subspaces in the first book, and they start from a matrix $\mathbf{A}$.  Each row of $\mathbf{A}$ is a vector in $n$-dimensional space.  When the matrix has $m$ rows, each column is a vector in $m$-dimensional space.  The crucial operation in linear algebra is to take ***linear combinations of column vectors***.  This is exactly the result of matrix-vector multiplication.  *$\mathbf{Ax}$ is a combination of the columns of $\mathbf{A}$*.

When we take *all* combinations $\mathbf{A}\mathbf{x}$ of the column vectors, we get the *column space*.  If this space includes the vector $\mathbf{b}$, then we can solve the equation $\mathbf{Ax}=\mathbf{b}$  In Section 1.3 there are two specific examlpes of this.

### The start of the course

The equation $\mathbf{Ax} = \mathbf{b}$ uses the language of linear algebra right away.  The vector $\mathbf{Ax}$ is *a combination of the columns of $\mathbf{A}$*.  The equation is asking for *a combination that produces $\mathbf{b}$.*  The solution vector $\mathbf{x}$ comes at three levels and they are all important:

1. ***Direct solution*** to find $\mathbf{x}$ by forward elimination and back substitution.

2. ***Matrix solution*** using the inverse matrix: $\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$ (if $\mathbf{A}$ has an inverse).

3. ***Particular solution*** (to $\mathbf{Ay} = \mathbf{b}$) plus ***nullspace solution*** (to $\mathbf{Az} = \mathbf{0}$).

That vector space solution $x = y + z$ is shown on the cover of the book.  Direct elimination is the most frequently used algorithm in scientific computing.  The matrix $\mathbf{A}$ becomes triangular - then solutions come quickly.  The speed of every new computer is tested on $\mathbf{Ax} = \mathbf{b}$: pure linear algebra.  But even a supercomputer doesn't want the inverse matrix: *too slow*.  Inverses give the simplest formula $\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$ but not the top speed.  And everyone must know that determinants are even slower - there is no way a linear algebra course should begin with formulas for the determinant of an $n$ by $n$ matrix.  Those formulas have a place, but not first place.

## Chapter 1: Introduction to vectors

The heart of linear algebra is in two operations - both with vectors.  We add vectors to get $\mathbf{v} + \mathbf{w}$.  We multiply them by numbers $c$ and $d$ to get $c\mathbf{v}$ and $d\mathbf{w}$.  Combining these two operations gives the ***linear combination*** $c\mathbf{v} + d\mathbf{w}$.

Linear comibations are all-important in this subjejct!  Sometimes we want one particular combination, and other times we want *all the combinations*.

The vectors $c\mathbf{v}$ lie along a line.  When $\mathbf{w}$ is not on that line, **the combinations $c\mathbf{v} + d\mathbf{w}$ fll the whole two-dimensional plane.**  

### 1.1 Vectors and linear combinations

> You can't add apples and oranges.

In a strange way, this is the reason for vectors.  We have two separate numbers $v_{1}$ and $v_{2}$.  That pair produces a **two-dimensional vector $\mathbf{v}$:**

$$
\mathbf{v} = 
\begin{bmatrix}
v_{1}\\
v_{2}
\end{bmatrix}
$$

We write $\mathbf{v}$ as a ***column***, not as a row.  Even if we don't add $v_1$ to $v_2$, we do ***add vectors***.  The first components of $\mathbf{v}$ and $\mathbf{w}$ stay separate from the second components:

$$
\mathbf{v} = 
\begin{bmatrix}
v_1\\
v_2
\end{bmatrix}
\quad
\text{and}
\quad
\mathbf{w} =
\begin{bmatrix}
w_1\\
w_2
\end{bmatrix}
\quad
\text{add to}
\quad
\mathbf{v} + \mathbf{w} = 
\begin{bmatrix}
v_1 + w_1\\
v_2 + w_2\\
\end{bmatrix}
$$

The other basic operation is *scaler multiplication*.  Vectors can be multipled by 2 or -1 or by any number $c$.  To find $\mathbf{v}$, multiply each component of $\mathbf{v}$ by 2:

$$
2\mathbf{v} = 
\begin{bmatrix}
2v_1\\
2v_2
\end{bmatrix}
=
\mathbf{v} + \mathbf{v}
$$

Notice that ths sum of $-\mathbf{v}$ and $\mathbf{v}$ is the zero vector.  This is **0**, which is not the same as the number zero!  The vector zero has components 0 and 0.

Linear algebra is built on these operations of $\mathbf{v} + \mathbf{w}$ and $c\mathbf{v}$ and $d\mathbf{w}$ - ***adding vectors and multiplying by scalers***.

#### Linear combinations

<div class="alert alert-block alert-info">
<b>The sum of $c\mathbf{v}$ and $d\mathbf{w}$ is a linear combination $c\mathbf{v} + d\mathbf{w}$</b>
</div>

Four special linear combinations are: sum, difference, zero, and a scaler multiple:

$$
\begin{align}
1\mathbf{v} + 1\mathbf{w} & = \text{sum of vectors} \\
1\mathbf{v} - 1\mathbf{w} & = \text{difference of vectors} \\
0\mathbf{v} + 0\mathbf{w} & = \text{zero vector} \\
c\mathbf{v} + 0\mathbf{w} & = \text{vector }c\mathbf{v}\text{ int he direction of }\mathbf{v}
\end{align}
$$

The zero vector is always a possible combination.  Every time we see a "space" of vectors, that zero vector will be included.  We have three ways to describe $\mathbf{v}$:

<div class="alert alert-block alert-info">
<b>Represent a vector $\mathbf{v}$ as two numbers, arrow from (0, 0), or a point in the plane</b>
</div>
 
Note that $\mathbf{w} + \mathbf{v}$ gives the same answer as $\mathbf{v} +\mathbf{w}$.

#### Vectors in three dimensions

<div class="alert alert-block alert-info">
<b>From now on $\mathbf{v} = \begin{bmatrix}1\\1\\-1\end{bmatrix}$ is also written as $\mathbf{v} = (1, 1, -1)$.</b>
</div>

This is simply to save space.

In three dimensions, $\mathbf{v} + \mathbf{w}$ is still found a component at a time.  A typical linear combination of three vectors in three dimensional space is $\mathbf{u} + 4\mathbf{v} - 2\mathbf{w}$:

$$
\begin{bmatrix}
1\\
0\\
3
\end{bmatrix}
+ 4
\begin{bmatrix}
1\\
2\\
1
\end{bmatrix}
- 2
\begin{bmatrix}
2\\
3\\
-1
\end{bmatrix}
=
\begin{bmatrix}
1\\
2\\
9
\end{bmatrix}
$$

#### The important questions

For one vector $\mathbf{u}$, the only linear combinations are $c\mathbf{u}$.  For two vectors, the combinations are $c\mathbf{u} + d\mathbf{v}$.  For three vectors, the combinations are $c\mathbf{u} + d\mathbf{v} + e\mathbf{w}$.  Will you take the big step from *one* combination to ***all combinations***?  Every $c$ and $d$ and $e$ are allowed.  SUppose the vectors $\mathbf{u}$, $\mathbf{v}$, $\mathbf{w}$ are in three-dimensional space:

1. What is the picture of *all* combinations of $c\mathbf{u}$?
<br>
<br>
2. What is the picture of *all* combinations of $c\mathbf{u} + d\mathbf{v}$?
<br>
<br>
3. What is the picture of *all* combinations of $c\mathbf{u} + d\mathbf{v} + e\mathbf{w}$?

The answers depend on the particular vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{w}$.  If they were zero vectors, then every combination would be zero.  If they are typical nonzero vectors, here are the three answers:

1. The combinations $c\mathbf{u}$ fill a ***line through*** (0, 0, 0).
<br>
<br>
2. The combinations $c\mathbf{u} + d\mathbf{v}$ fill a ***plane through*** (0, 0, 0).
<br>
<br>
3. The combinations $c\mathbf{u} + d\mathbf{v} + e\mathbf{w}$ fill ***three-dimensional space***.

#### Review of key ideas

1. A vector $\mathbf{v}$ in two-dimensional space has two components $v_1$ and $v_2$.
<br>
<br>
2. $\mathbf{v} + \mathbf{w} = (v_1 + w_1, v_2 + w_2)$ and $c\mathbf{v} = (cv_1, cv_2)$ are found a component at a time.
<br>
<br>
3. A linear combination of three vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{w}$ is $c\mathbf{u} + d\mathbf{v} + e\mathbf{w}$.
<br>
<br>
4. Take *all* linear combinations of $\mathbf{u}$, or $\mathbf{u}$ and $\mathbf{v}$, or $\mathbf{u}$, $\mathbf{v}$, $\mathbf{w}$.  In three dimesions, those combinations typically fill a line, then a plane, then the whole space $\mathbb{R}^3$.
