#### Problem 1 calculations

From [Strang, Gilbert. Introduction to Linear Algebra. Wellesley, Ma, Cambridge Press, 2016][1], given $\mathbf{u}\text{, }\mathbf{v}\text{, }\mathbf{w}$ which occupy space in $\mathbb{R}^n$:

- $\mathbf{u}\text{, }\mathbf{v}\text{, }\mathbf{w}$ are linearly independent if no other linear combination except $0\mathbf{u} + 0\mathbf{v} + 0\mathbf{w} = \mathbf{0}$
<br>
<br>
- Otherwise the vectors are linearly dependent

<br>
Given
$
\qquad
\begin{bmatrix}
  \begin{pmatrix}
    1\\
    2\\
    3
  \end{pmatrix}
  \text{,}
  \begin{pmatrix}
    2\\
    1\\
    0
  \end{pmatrix}
  \text{,}
  \begin{pmatrix}
    1\\
    5\\
    9
  \end{pmatrix}
\end{bmatrix}
$
<br>
<br>
let: 
$
\qquad
\mathbf{u} = \begin{pmatrix}1\\2\\3\end{pmatrix}\text{,}
\qquad
\mathbf{v} = \begin{pmatrix}2\\1\\0\end{pmatrix}\text{,}
\qquad
\mathbf{w} = \begin{pmatrix}1\\5\\9\end{pmatrix}\text{,}
\qquad
\mathbf{b} = \begin{pmatrix}0\\0\\0\end{pmatrix}\text{,}
\qquad
\text{and }
c\text{, }d\text{, }e \in{} \mathbb{R}
$
<br>

The set of all linear combinations can be written as $c\mathbf{u} + d\mathbf{v} + e\mathbf{w}$

Since $\mathbf{Ax}$ is the linear combination of the columns of $\mathbf{A}$, let:

$$
\mathbf{Ax} =
\begin{bmatrix}
  1 & 2 & 1 \\
  2 & 1 & 5 \\
  3 & 0 & 9 
\end{bmatrix}
\begin{bmatrix}
  c\\
  d\\
  e
\end{bmatrix}
$$

Setting $\mathbf{Ax} = \mathbf{b}$:

$$
\begin{bmatrix}
  1 & 2 & 1 \\
  2 & 1 & 5 \\
  3 & 0 & 9 
\end{bmatrix}
\begin{bmatrix}
  c\\
  d\\
  e
\end{bmatrix}
=
\begin{bmatrix}
  0\\
  0\\
  0
\end{bmatrix}
$$

And performing Guassian Elimination:

$$
\left[\begin{array}{rrr|r}
  1 & 2 & 1 & 0 \\
  2 & 1 & 5 & 0 \\
  3 & 0 & 9 & 0
\end{array}\right]
$$

$$
\left[\begin{array}{rrr|r}
  1 & 2 & 1 & 0 \\
  0 & -3 & 3 & 0 \\
  3 & 0 & 9 & 0
\end{array}\right]
$$

$$
\left[\begin{array}{rrr|r}
  1 & 2 & 1 & 0 \\
  0 & -3 & 3 & 0 \\
  0 & -6 & 6 & 0
\end{array}\right]
$$

$$
\left[\begin{array}{rrr|r}
  1 & 2 & 1 & 0 \\
  0 & 0 & 0 & 0 \\
  0 & 1 & -1 & 0
\end{array}\right]
$$

Row reduced echelon form:

$$
\left[\begin{array}{rrr|r}
  1 & 0 & 3 & 0 \\
  0 & 1 & -1 & 0 \\
  0 & 0 & 0 & 0
\end{array}\right]
$$

$$
\begin{bmatrix}
  1 & 0 & 3 \\
  0 & 1 & -1 \\
  0 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
  c \\
  d \\
  e
\end{bmatrix}
=
\begin{bmatrix}
  c + 3e \\
  d - e \\
  0
\end{bmatrix}
=
\begin{bmatrix}
  0 \\
  0 \\
  0
\end{bmatrix}
$$

$$
\begin{align}
d & = e \\
c & = -\frac{1}{3}e
\end{align}
$$

Let $c = 3$ then:

$$
3
\begin{bmatrix}
  1 \\
  2 \\
  3 \\
\end{bmatrix}
- 1
\begin{bmatrix}
  2 \\
  1 \\
  0
\end{bmatrix}
- 1
\begin{bmatrix}
  1 \\
  5 \\
  9
\end{bmatrix}
=
\begin{bmatrix}
  3 - 2 - 1 \\
  6 - 1 - 5 \\
  9 - 0 - 9
\end{bmatrix}
=
\begin{bmatrix}
  0 \\
  0 \\
  0
\end{bmatrix}
$$
<br>
<div style="border:solid; margin: auto; text-align: center; padding: 1em">
<b>Therefore the set of vectors are linearly dependent</b>
</div>
